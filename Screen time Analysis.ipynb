```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('C:/Users/Windows10/Downloads/Screentime-App-Details.csv')
data
```

```python
data.info()
```

```python
Date=data.sort_values(by='Date',ascending=True)
Date
```

```python
plt.figure(figsize=(10,8))
sns.barplot(Date,x='Date',y='Usage',hue='App')
plt.xticks(rotation = 75)
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.catplot(Date,kind='bar',y='Usage',x='App')
plt.xticks(rotation = 75)
plt.show()
```

## Take Away
### 1. Most used app after analysing the data is Whatsapp.

```python
plt.figure(figsize=(10,8))
sns.barplot(Date,x='Date',y='Times opened',hue='App')
plt.xticks(rotation = 75)
plt.show()
```
